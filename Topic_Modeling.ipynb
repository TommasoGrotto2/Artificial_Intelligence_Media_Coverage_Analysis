```python
# -------------------
# 1. Install libraries
# -------------------
!pip install -q gensim spacy pyLDAvis
!python -m spacy download en_core_web_sm  # for English lemmatization

# -------------------
# 2. Import libraries
# -------------------
import pandas as pd
import gensim
from gensim import corpora
import spacy
from nltk.corpus import stopwords
import nltk
import pyLDAvis.gensim
import pyLDAvis

nltk.download('stopwords')

# -------------------
# 3. Load data
# -------------------
# Change to your file path
df = pd.read_csv("/content/drive/MyDrive/Mediacloud/UK/artificial_intelligence_uk_cleaned_2024-06-01_2025-06-01_frames_sentiments_newsplease_mediacloud_gemini_2-5.csv")

# We'll use the maintext column
texts = df["maintext"].dropna().astype(str).tolist()

# -------------------
# 4. Preprocess text
# -------------------
nlp = spacy.load("en_core_web_sm", disable=["parser", "ner"])
stop_words = set(stopwords.words("english"))

def preprocess(text):
    doc = nlp(text.lower())
    tokens = [
        token.lemma_ for token in doc
        if token.is_alpha and token.text not in stop_words and len(token.text) > 3
    ]
    return tokens

processed_texts = [preprocess(t) for t in texts]

# -------------------
# 5. Create dictionary & corpus
# -------------------
dictionary = corpora.Dictionary(processed_texts)
dictionary.filter_extremes(no_below=5, no_above=0.5)  # remove very rare/common words

corpus = [dictionary.doc2bow(text) for text in processed_texts]

# -------------------
# 6. Train LDA model
# -------------------
num_topics = 10  # you can tune this
lda_model = gensim.models.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    random_state=42,
    passes=10,
    alpha='auto'
)

# -------------------
# 7. Display topics
# -------------------
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}:\n{topic}\n")

# -------------------
# 8. Visualize topics
# -------------------
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim.prepare(lda_model, corpus, dictionary)
vis
```

```python
# -------------------
# 2. Import libraries
# -------------------
import pandas as pd
from bertopic import BERTopic

# -------------------
# 3. Load dataset
# -------------------
# Change to your file path
df = pd.read_csv("/content/drive/MyDrive/Mediacloud/UK/artificial_intelligence_uk_cleaned_2024-06-01_2025-06-01_frames_sentiments_newsplease_mediacloud_gemini_2-5.csv")

import re
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))

def clean_text(text):
    # Lowercase
    text = text.lower()
    # Remove non-alphabetical characters
    text = re.sub(r'[^a-z\s]', ' ', text)
    # Remove stopwords
    text = " ".join([w for w in text.split() if w not in stop_words])
    return text

texts = df["maintext"].dropna().astype(str).apply(clean_text).tolist()

# -------------------
# 4. Create and fit BERTopic model
# -------------------
#topic_model = BERTopic(language="multilingual", verbose=True)
topic_model = BERTopic(language="english", min_topic_size=30, verbose=True)
topics, probs = topic_model.fit_transform(texts)

# -------------------
# 5. Inspect topics
# -------------------
# Get topic info
topic_info = topic_model.get_topic_info()
print(topic_info.head(10))  # top 10 topics

# -------------------
# 6. See keywords for a specific topic
# -------------------
topic_id = 0
print(f"Topic {topic_id} keywords:\n", topic_model.get_topic(topic_id))

# -------------------
# 7. Visualizations
# -------------------
topic_model.visualize_topics()
topic_model.visualize_barchart(top_n_topics=10)
topic_model.visualize_hierarchy()
```
